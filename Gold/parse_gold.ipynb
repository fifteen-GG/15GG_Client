```python
import json
```

```python
with open('item.json') as item_list:
    json_object = json.load(item_list)

item_dict = dict()
data = json_object['data']
```

```python
for item_id in data.keys():
    item_dict[item_id] = {'name' : data[item_id]['name'], 'gold' : data[item_id]['gold']['total']}
```
